```python
def vowels(s):
    vows = ['a', 'e', 'i','o','u']
    d={}
    for each in s:
        if each in vows:
            if each in d.keys():
                d[each] = d[each] +1
            else:
                d[each]=1
    return d
```

```python
vowels('aeiou')
```

```python
from bs4 import BeautifulSoup
import mechanize
```

```python
br = mechanize.Browser()
br.set_handle_robots(False)

ua = 'Mozilla/5.0 (X11; Linux x86_64; rv:18.0) Gecko/20100101 Firefox/18.0 (compatible;)'
br.addheaders = [('User-Agent', ua), ('Accept', '*/*')]
response2 = br.open("http://www.google.com/patents/sitemap/en/Sitemap.html")

result = response2.read()
soup = BeautifulSoup(result)
soup
```

```python
soup.title
```

```python
element = soup.findAll('a',{'class': 'subcategory-link'})
element
```

extracting strings in atags

```python
a=[]
for each in element:
    a.append(each.string)
```

```python
state = sample.groupby('State').agg({'Revenue': sum}).reset_index()
state
```

sort a/c to top revenue of states

```python
sort_state = state.sort_values(by='Revenue',ascending=False)
sort_state.head(5)
```

```python
sort_state.tail(5)
```

sort a/c district

```python
sort_dist = sample.groupby(['State','District']).agg({'Revenue': sum}).reset_index()
sort_dist
```

```python
sort_dist[sort_dist['State']=='Karnataka'].sort_values(by='Revenue',ascending =False)
```
